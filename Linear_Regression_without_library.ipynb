```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([6,7,8,9,10])

mx=x.mean()
my=y.mean()

cov_xy=np.mean((x-mx)*(y-my))
var_x=np.mean((x-mx)**2)

m=cov_xy/var_x
c=my-m*mx

y_reg=str(m)+"x"+"+"+str(c)
print(y_reg)
```
