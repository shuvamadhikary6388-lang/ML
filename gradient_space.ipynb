```python
import numpy as np
import matplotlib.pyplot as plt
B1=[]
B0=[]
def gradient_descent(x, y, lr=0.1, n_iters=10):
    m = len(x)
    b0 = 0  # intercept
    b1 = 0  # slope
    for i in range(n_iters):
        y_pred = b0 + b1 * x
        db0 = (-1/m) * np.sum(y - y_pred)
        db1 = (-1/m) * np.sum(x * (y - y_pred))
        b0 = b0 - lr * db0
        b1 = b1 - lr * db1
        if i % 100 == 0:
            cost = (1/(2*m)) * np.sum((y - y_pred)**2)
            print(f"Iter {i}: Cost={cost:.4f}, b0={b0:.4f}, b1={b1:.4f}")
        B0.append(b0)
        B1.append(b1)
    return b0, b1
x = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([5,6,10,13,11], dtype=float)

b0, b1 = gradient_descent(x, y, lr=0.01, n_iters=1000)
print("b0 =", b0)
print("b1 =", b1)

import matplotlib.pyplot as plt
plt.plot(B0, label='B0')
plt.plot(B1, label='B1')
plt.legend()
plt.show()
```
