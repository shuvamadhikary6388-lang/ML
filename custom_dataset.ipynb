```python
import pandas as pd
import random
student=[[0]*5 for i in range(20)]
for i in range(20):
  for j in range(5):
    student[i][j]=random.randint(1,1000)
head=['c1','c2','c3','c4','c5']
df=pd.DataFrame(student,columns=head)
df.to_csv("student.csv",index=False)
print(df)
```
